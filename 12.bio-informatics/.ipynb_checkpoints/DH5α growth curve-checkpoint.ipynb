```python
def plot_growth_curve(time, rpm_150, rpm_230, group, legend_1, legend_2, x_label, y_label):
    rpm_150_np = np.array(rpm_150)
    rpm_230_np = np.array(rpm_230)
    df1 = pd.DataFrame(rpm_150_np,
                   index=time,
                   columns=[legend_1])
    df2 = pd.DataFrame(rpm_230_np,
                   index=time,
                   columns=[legend_2])
    df_result = df1.join(df2)
    
#     df_result.plot()
    
    f1 = plt.figure(figsize=(12, 11))
#     plt.ylim(0.0, 3.0)
    plt.plot(time, rpm_150,  color='black', marker='o',\
         linestyle='-',  linewidth=2, markersize=12,  alpha=.5)
    plt.plot(time, rpm_230,  color='black', marker='*',\
         linestyle='--',  linewidth=2, markersize=12,  alpha=.5)
    plt.title("DH5Î± growth curve-group "+group)
    plt.rc('font', size=20)
    plt.rc('xtick', labelsize=20)
    plt.rc('ytick', labelsize=20)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, lw = 2, ls = '--', c = '.85')
    plt.legend([legend_1, legend_2])
    plt.show()
```

```python
def calculate_mu(time, rpm_150, rpm_230):
    dt = []
    dOD_rpm_150 = []
    dOD_rpm_230 = []
    growth_rate_150rpm = []
    growth_rate_230rpm = []
    mu_per_hours_150rpm = []
    mu_per_hours_230rpm = []
    doubling_time_per_hours_150rpm = []
    doubling_time_per_hours_230rpm = []
    
    for time_idx in range(len(time)-1):
        dt.append(time[time_idx+1] - time[0])
    print("dt: {}".format(dt))
    for rpm_idx in range(len(rpm_150)-1):
        dOD_rpm_150.append(rpm_150[rpm_idx+1] - rpm_150[rpm_idx])
    print("dOD_rpm_150: {}".format(dOD_rpm_150))
    for rpm_idx in range(len(rpm_230)-1):
        dOD_rpm_230.append(rpm_230[rpm_idx+1] - rpm_230[rpm_idx])
    print("dOD_rpm_230: {}".format(dOD_rpm_230))
    
    for i in range(len(dt)): #dOT/dt
        growth_rate_150rpm.append(dOD_rpm_150[i] / dt[i])
        growth_rate_230rpm.append(dOD_rpm_230[i] / dt[i])
    print("growth_rate_150rpm: {}".format(growth_rate_150rpm))
    print("growth_rate_230rpm: {}".format(growth_rate_230rpm))
    
    for i in range(len(dOD_rpm_150)): # (dOT/dt)/OD
        mu_per_hours_150rpm.append(growth_rate_150rpm[i] / rpm_150[i])
    print("mu_per_hours_150rpm: {}".format(mu_per_hours_150rpm))
    
    for i in range(len(dOD_rpm_230)): # (dOT/dt)/OD
        mu_per_hours_230rpm.append(growth_rate_230rpm[i] / rpm_230[i])
    print("mu_per_hours_230rpm: {}".format(mu_per_hours_230rpm))
    
    for i in range(len(mu_per_hours_150rpm)):
        doubling_time_per_hours_150rpm.append(math.log(2, math.e) / mu_per_hours_150rpm[i])
    print("doubling_time_per_hours_150rpm: {}".format(doubling_time_per_hours_150rpm))
    
    for i in range(len(mu_per_hours_230rpm)):
        doubling_time_per_hours_230rpm.append(math.log(2, math.e) / mu_per_hours_230rpm[i])
    print("doubling_time_per_hours_150rpm: {}".format(doubling_time_per_hours_230rpm))
        
    dOD_rpm_150.insert(0, 0)
    dOD_rpm_230.insert(0, 0)
    growth_rate_150rpm.insert(0, 0)
    growth_rate_230rpm.insert(0, 0)
    mu_per_hours_150rpm.insert(0, 0)
    mu_per_hours_230rpm.insert(0, 0)
    doubling_time_per_hours_150rpm.insert(0, 0)
    doubling_time_per_hours_230rpm.insert(0, 0)
    
    return mu_per_hours_150rpm, mu_per_hours_230rpm
```
