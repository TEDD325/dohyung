# 모범답안 1

```python
import requests
req = requests.get('http://cse.koreatech.ac.kr')
source = req.text

text = "" #순수 텍스트 식별자
while(1) :      
    source = source[source.find('<'):] #'<'를 찾는다
    if source.find("<!--") == 0 : source = source[source.find("-->")+3:] #주석 내용 건너뜀
    elif source.find("<script>") == 0 : source = source[source.find("</script>")+9:] #자바스크립트 내용 건너뜀   
    else :
        source = source[source.find('>')+1:] #'>'를 찾는다
        for ch in source :    #'>' 이후부터 한글자씩 가져온다
            if ch == '<' : break    # 다시 '<' 문자 등장하면 종료
            elif ch == '\t' or ch == '\n' : continue    #탭,개행 문자는 순수 텍스트에 포함x
            else : text += ch   #순수 텍스트 추가     
        if text.endswith(" ") == False : text += " " #공백으로 분리하기 위해 순수 텍스트 추출 후 뒤에 공백문자 추가
    if(source.find('<') == -1):break #'<' 문자가 더 이상 없는 경우 무한루프종료 

# print(text)   

list = text.split() #추출한 순수 텍스트 문자열을 공백 단위로 분리
print (list)

print (len(list)) #분리된 단어 수 출력
```

# 모범답안 2

```python
import requests
req = requests.get('http://cse.koreatech.ac.kr')
source = req.text
startHTML = False # html 태그가 열린 상태인지 여부
textList = []
string = ''
for i in range(0, len(source)) :
    if startHTML == True : # html태그가 열렸다면
        if source[i] != '>' : continue # html태그가 닫히기 전까지 무시하고 진행
        else : startHTML = False # 닫혔다면 닫힌상태로 표시.
    elif source[i] == '<' : startHTML = True # 열렸다면 열린상태로 표시
    elif source[i] == ' ' or source[i] == '\n' or source[i] == '' or source[i] == '\t' or source[i] == '\r' : # 닫혀있는 상태라면 공백문자인지 검사
        if string == '' : continue # 단어가 없다면 무시하고 진행
        else :
            textList.append(string) # 걸러낸 단어가 있다면 추출
            string = ''
    else : string = string + source[i] # 단어 글자단위로 걸러내기
textList.append(string)
print(len(textList))
print(textList)
# print(source)
```
