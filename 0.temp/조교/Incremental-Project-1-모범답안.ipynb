# 모범답안 1

```python
import requests
req = requests.get('http://cse.koreatech.ac.kr')
source = req.text

text = "" #순수 텍스트 식별자
while(1) :      
    source = source[source.find('<'):] #'<'를 찾는다
    if source.find("<!--") == 0 : source = source[source.find("-->")+3:] #주석 내용 건너뜀
    elif source.find("<script>") == 0 : source = source[source.find("</script>")+9:] #자바스크립트 내용 건너뜀   
    else :
        source = source[source.find('>')+1:] #'>'를 찾는다
        for ch in source :    #'>' 이후부터 한글자씩 가져온다
            if ch == '<' : break    # 다시 '<' 문자 등장하면 종료
            elif ch == '\t' or ch == '\n' : continue    #탭,개행 문자는 순수 텍스트에 포함x
            else : text += ch   #순수 텍스트 추가     
        if text.endswith(" ") == False : text += " " #공백으로 분리하기 위해 순수 텍스트 추출 후 뒤에 공백문자 추가
    if(source.find('<') == -1):break #'<' 문자가 더 이상 없는 경우 무한루프종료 

# print(text)   

list = text.split() #추출한 순수 텍스트 문자열을 공백 단위로 분리
print (list)

print (len(list)) #분리된 단어 수 출력
```

# 모범답안 3

```python
import requests

req = requests.get('http://cse.koreatech.ac.kr')

source = req.text #컴공홈페이지의 소스를 받아온다.

a="" #<>제외한 문자열을 넣어줄 a를 정의
i=0 #i를 0으로 정의
while (i < len(source)): #소스크기 만큼 반복하는 반복문
    if(source[i] == '<'): #요소가 <이 나올 경우
        i=i+1 #우선 다음 요소로 넘어가서
        while(source[i] != '>'): #요소가 >이 나올때까지
            i=i+1 #계속해서 다음 요소로 넘어간다.
        a=a+' ' # >이 나와서 while문을 탈출하면 공백을 a문자열에 넣어준다.
        i=i+1 # 다음 요소로 넘어감
    else:
        a=a+source[i] # 요소가 <가 아닐 경우에는 a문자열에 넣어준다.
        i=i+1 #다음 요소로 넘어감
b=a.split() #공백 단위로 a의 문자열을 나누어 b에 넣어준다. (단어를 구분함)
for j in range(0,len(b)): #b의 크기만큼
    print(b[j],end='\n') #단어들을 출력
print()
print()
print("총 단어의 개수는 {0}개입니다".format(len(b))) #단어개수 출력
```

# 모범답안 4

```python
import requests
import re
req = requests.get('http://cse.koreatech.ac.kr')
source = req.text

#print(source)

def remove_tags(doc):
    text = re.sub("<script[\s\S]+?/script>", " ", doc) #<script> 태그 지우기
    text = re.sub("<.*?>", " ", text) #<>태그 지우기
    text = re.sub("[\n\t]", " ", text) #엔터 및 탭 지우기
    text = text.strip() # 공백 지우기
    return text

text = remove_tags(source).split()
print(len(text))
```
