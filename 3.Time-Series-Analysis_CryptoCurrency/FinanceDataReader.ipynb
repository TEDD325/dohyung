```python
df_krx = fdr.StockListing('KOSPI')
df_krx
```

```python
def find_symbol(stock):
    for idx in range(len(df_krx['Name'])):
        if df_krx['Name'][idx] == str(stock):
            print(df_krx['Symbol'][idx])
    #         print("True")
```

```python
stock = '삼성전자'
find_symbol(stock)
```

```python
stock = 'LG전자'
find_symbol(stock)
```

```python
stock = 'NAVER'
find_symbol(stock)
```

```python
stock = '카카오'
find_symbol(stock)
```
