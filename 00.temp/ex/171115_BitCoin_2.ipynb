```python
import time
def crawlingFromPoloneix(coinType):
    transactionData = []
    checkData = ''
    checkUnit = ''
    checkprice = ''
    checkType = ''
    
    url = 'https://poloniex.com/public?command=returnTradeHistory&currencyPair=BTC_'+coinType+'&limit=1'
    urlTicker_poloniex = urllib.request.urlopen(url)
    readTicker_poloniex = urlTicker_poloniex.read()
    t_poloniex = json.loads(readTicker_poloniex)
    checkData = t_poloniex[0]['date']
    checkUnit = t_poloniex[0]['amount']
    checkprice = t_poloniex[0]['rate']
    while(True):
        time.sleep(1)

        url = 'https://poloniex.com/public?command=returnTradeHistory&currencyPair=BTC_'+coinType+'&limit=1'
        urlTicker_poloniex = urllib.request.urlopen(url)
        readTicker_poloniex = urlTicker_poloniex.read()
        t_poloniex = json.loads(readTicker_poloniex)
        transactionData.append(t_poloniex)
        if (transactionData[len(transactionData)-1][0]['date'] != checkData 
                or transactionData[len(transactionData)-1][0]['amount'] != checkUnit  
                or transactionData[len(transactionData)-1][0]['rate'] != checkprice
                or transactionData[len(transactionData)-1][0]['type'] != checkType):
            print(t_poloniex, end="\n\n")
            checkData = t_poloniex[0]['date']
            checkUnit = t_poloniex[0]['amount']
            checkprice = t_poloniex[0]['rate']
            checkType = t_poloniex[0]['type']
            
            
def crawlingFromBithumb(coinType):
    new_dict = {}
    new_list = []
    transactionData = []
    checkData = ''
    checkUnit = ''
    checkprice = ''
    checkType = ''
    url = 'https://api.bithumb.com/public/recent_transactions/'+coinType+'?count=1'
    urlTicker_bithumb = urllib.request.urlopen(url)
    readTicker_bithumb = urlTicker_bithumb.read()
    t_bithumb = json.loads(readTicker_bithumb)
    checkData = t_bithumb['data'][0]['transaction_date']
    checkUnit = t_bithumb['data'][0]['units_traded']
    checkprice = t_bithumb['data'][0]['price']
    while(True):
        time.sleep(1)

        url = 'https://api.bithumb.com/public/recent_transactions/'+coinType+'?count=1'
        urlTicker_bithumb = urllib.request.urlopen(url)
        readTicker_bithumb = urlTicker_bithumb.read()
        t_bithumb = json.loads(readTicker_bithumb)
        transactionData.append(t_bithumb)
        if (transactionData[len(transactionData)-1]['data'][0]['transaction_date'] != checkData 
                or transactionData[len(transactionData)-1]['data'][0]['units_traded'] != checkUnit  
                or transactionData[len(transactionData)-1]['data'][0]['price'] != checkprice
                or transactionData[len(transactionData)-1]['data'][0]['type'] != checkType):
            print(t_bithumb, end="\n\n")
            checkData = t_bithumb['data'][0]['transaction_date']
            checkUnit = t_bithumb['data'][0]['units_traded']
            checkprice = t_bithumb['data'][0]['price']
            checkType = t_bithumb['data'][0]['type']
```
