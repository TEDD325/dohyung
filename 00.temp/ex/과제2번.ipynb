> - 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
>   - 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
>   - 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
>   - 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
>   - 반환하는 리스트는 정렬이 되어 있어야 한다.
>   - 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
>     - list_union([1, 2, 3], [1, 2, 4])
>       - [1, 2, 3, 4]
>     - list_union([-10, -5, 0, -1], [100, 9, 0, 9])
>       - [-10, -5, -1, 0, 9, 100]
>     - list_union([0, 1, 2], [0, 1, 2])
>       - [0, 1, 2]
>   - [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
>   - [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

````python
list__a = [1,2,3.1]
list__b = [1,2,4]

check_int = list__a + list__b
for item in check_int:
    if type(int()) != type(item):
        print("<warning> '{:2.1f}' is not an integer".format(item))
    else:
        print("{:d} is an integer..".format(item))
````

````python
def list_union(lista, listb):
    list__result = lista
    check_int = lista + listb
    for item in check_int:
        if type(int()) != type(item):
            print("<warning> '{:2.1f}' is not an integer".format(item))
        #else:
            #print("{:3d} is an integer..".format(item))
    
    for i in range(len(lista)):
        if (listb[i] in lista): # x in l
            continue
        else:
            list__result.append(listb[i])
        list__result = sorted(list__result)
    print(list__result)
````

````python
def wordCount_in_WebPage(pageContent):
    test_dict = {}
    words = []
    test_string = ''
    string = pageContent
    
    while '</html>' != (string[string.index('<'):string.index('>')+1]):
        string = string[string.index('>')+1:]
        word = string[:string.index('<')]
        words.append(word)
        
    for i in range(len(words)):
        test_string = test_string + ' ' + words[i]
        
    words = test_string.split()

    for i in range(len(words)): # 초기화
        test_dict[words[i]] = 0

    
    for i in range(len(words)):
        test_dict[words[i]] = test_dict[words[i]]+1
    # print(test_dict)
    
    print("--word completed--")
    
    return test_dict
````

> 26부터 시작한다. 2+6 = 8이다. 새로운 숫자는 68이다. 6+8 = 14이다. 새로운 숫자는 84이다. 8+4 = 12이다. 새로운 숫자는 42이다. 4+2 = 6이다. 새로운 숫자는 26이다.<br>
> 위의 예는 4번만에 원래 숫자로 돌아올 수 있다. 따라서 26의 사이클의 길이는 4이다.<br>
