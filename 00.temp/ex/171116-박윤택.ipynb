```python
from urllib import request

url = 'http://www.nytimes.com'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')

#추가되는 부분 - 1

p=0
a=source.strip()
s=a.replace(a[a.find('<script',p):a.find('</script>',p)+9], '')
s=s.replace(s[s.find('<style',p):s.find('</style>',p)+8], '')

for x in range(len(source.splitlines())):
    s=s.replace(s[s.find('<script',p):s.find('</script>',p)+9], '')
    s=s.replace(s[s.find('<style',p):s.find('</style>',p)+8], '')


l=s.split()
check = 0
one = 0
u = []

for x in range(len(l)):
    for y in range(len(l[x])):
        if(l[x][y] == '<'):
            check = 1
        if(check == 0 and one == 0):
            u.append(l[x])
            one = 1

        if(l[x][y] == '>'):
            check = 0

    one = 0

result=[]

for x in range(len(u)):
    if((u[x].find('<') == -1 ) and (u[x].find('>') == len(u[x])-1)):
        continue
    else:       
        result.append(u[x])  

        
w=[]
start = 0
finish = 0

for x in range(len(result)):
    if(result[x].find('<') == 0):
        start = result[x].find('>') + 1
        if(result[x].find('<', 1) != -1):
            finish = result[x].find('<',1)
            w.append(result[x][start:finish])

        else:
            w.append(result[x][start:])

    elif(result[x].find('<') > 0 and result[x].find('<') < result[x].find('>')):
        finish = result[x].find('<')
        w.append(result[x][:finish])

    elif(result[x].find('<')==-1 and result[x].find('>')>=0):
        start = result[x].find('>')+1
        w.append(result[x][start:])

    elif(result[x].find('>') < result[x].find('<')):
        start = result[x].find('>')+1
        finish = result[x].find('<')
        w.append(result[x][start:finish])

    elif(result[x].find('<') > 0 and result[x].find('>') > 0):
        finish = result[x].find('<')
        w.append(result[x][:finish])
        start = result[x].find('>')+1
        w.append(result[x][start:])
    
    else:
        w.append(result[x])

    start = 0
    finish = 0
```
