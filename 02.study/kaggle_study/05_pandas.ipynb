## # DataFrame 생성

### - python dictionary ->  pandas.DataFrame
> <code>pd.DataFrame( python dict )</code>

```python
import pandas as pd
```

```python
# create a simple dataset of people
data_dict = {'Name': ["John", "Anna", "Peter", "Linda"],
        'Location' : ["New York", "Paris", "Berlin", "London"],
        'Age' : [24, 13, 53, 33]
       }

data_pandas = pd.DataFrame(data_dict)
print(data_pandas)
```

#### columns 명시

```python
# create a simple dataset of people
data_dict = {'Name': ["John", "Anna", "Peter", "Linda"],
        'Location' : ["New York", "Paris", "Berlin", "London"],
        'Age' : [24, 13, 53, 33]
       }

data_pandas = pd.DataFrame(data_dict, columns=tuple(data_dict.keys()))
print(data_pandas)
```

### - plotting

#### scatter matrix

```python
import pandas as pd
import mglearn
%matplotlib inline

from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("type:",type(iris_dataset))
'''IRIS 데이터: supervised learning 중 classification 문제에 해당'''

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'],
    iris_dataset['target'],
    test_size = 0.25, # default
    random_state = 0, # seed. 높은 재연성.
    shuffle = True # default
)

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

pd.plotting.scatter_matrix(
    iris_dataframe, 
    c=y_train, 
    figsize=(12, 12), 
    marker='o',
    cmap=mglearn.cm3
)
```
